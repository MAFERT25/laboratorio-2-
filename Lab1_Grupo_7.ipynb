```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

x = np.linspace(-5, 5, 100)
y = f(x)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Plot of a 2D Scalar Function")
plt.grid()
plt.show()
```
